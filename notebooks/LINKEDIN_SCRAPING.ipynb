```python
import json
import os

from supabase import create_client

url = "https://omiaednmxvekojyflhgi.supabase.co"
key = os.getenv("SUPABASE_KEY")
supabase = create_client(url, key)

FILE_PATH = "/content/dataset_linkedin-jobs-scraper_2025-11-29_13-20-49-548.json"

SOURCE_MAP = {
    "1": "https://www.linkedin.com/posts/bayer",
    "2": "https://www.linkedin.com/posts/amgen",
    "3": "https://www.linkedin.com/posts/gilead-sciences",
    "4": "https://www.linkedin.com/posts/regeneron-pharmaceuticals",
    "5": "https://www.linkedin.com/posts/biogen",
    "6": "https://www.linkedin.com/posts/modernatx",
    "7": "https://www.linkedin.com/posts/pfizer",
    "8": "https://www.linkedin.com/posts/johnson-&-johnson",
    "9": "https://www.linkedin.com/posts/roche",
    "10": "https://www.linkedin.com/posts/novartis",
    "11": "https://www.linkedin.com/posts/msd-global",
    "12": "https://www.linkedin.com/posts/basf",
    "13": "https://www.linkedin.com/posts/syngenta",
    "14": "https://www.linkedin.com/posts/corteva",
    "15": "https://www.linkedin.com/posts/merck-group",
    "16": "https://www.linkedin.com/posts/sanofi",
    "17": "https://www.linkedin.com/showcase/cortevabiologicals"
}

LOOKUP = {}

for sid, link in SOURCE_MAP.items():
    link = link.lower()
    if "linkedin.com/posts/" in link:
        key = link.split("linkedin.com/posts/")[1]
    elif "linkedin.com/showcase/" in link:
        key = link.split("linkedin.com/showcase/")[1]
    else:
        continue

    LOOKUP[key] = int(sid)

def normalize(name: str):
    return (
        name.lower()
        .replace("&", "and")
        .replace(" ", "-")
        .replace(",", "")
        .replace(".", "")
        .strip()
    )

with open(FILE_PATH, encoding="utf-8") as f:
    jobs = json.load(f)

print("Found", len(jobs), "job records")

id_counter = 51

for job in jobs:
    company_raw = job.get("companyName")
    if not company_raw:
        continue

    comp = normalize(company_raw)

    matched_source = None
    for key in LOOKUP:
        if comp.startswith(normalize(key)):
            matched_source = LOOKUP[key]
            break

    if not matched_source:
        continue

    date_posted = job.get("publishedAt")
    if not date_posted:
        continue

    row = {
        "id": id_counter,
        "data": job,
        "source_id": matched_source,
        "date_posted": date_posted
    }

    supabase.table("job").insert(row).execute()
    print(f"Inserted job ID={id_counter} for {company_raw} → source_id={matched_source}")

    id_counter += 1  # increment ID manually
```
