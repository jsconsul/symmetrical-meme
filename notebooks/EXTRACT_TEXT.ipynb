```python
import os

import pandas as pd
from supabase import create_client

url = "https://omiaednmxvekojyflhgi.supabase.co"
key = os.getenv("SUPABASE_KEY")
supabase = create_client(url, key)

response = supabase.table('post').select('*').execute()
all_posts_data = response.data

df_posts = pd.DataFrame(all_posts_data)
display(df_posts.head(3166))
```

```python
extracted_contents = []

for post in all_posts_data:
    post_id = post['id']
    post_data = post['data']
    extracted_text = post_data.get('text', '')

    extracted_contents.append({
        "post_id": post_id,
        "extracted_text": extracted_text
    })

for content in extracted_contents:
    print(f"Post ID: {content['post_id']}, Extracted Text: {content['extracted_text'][:100]}...")
```

```python
id_counter = 1
for content in extracted_contents:
    row = {
        "id": id_counter,
        "post_id": content["post_id"],
        "extracted_text": content["extracted_text"],
        "intent": "",
        "topics": [],
        "keywords": [],
        "entities": []
    }
    supabase.table("post_content").insert(row).execute()
    print(f"Inserted post_id: {content['post_id']} into post_content table.")
    id_counter+=1
```
